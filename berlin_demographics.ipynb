```python
# Retrieve dataset with German nationality
import pandas as pd
df_unmelted = pd.read_csv(r"C:\Users\aican\2025-Ironhack\GitHub\Unit4-2025-CW18-19\sql-database\berlin_demographics_unmelted.csv")
```

```python
df_unmelted
```

```python
# Melt the DataFrame
df_melted = df_unmelted.melt(
    id_vars=["Altersgruppen", "Deutsche"],
    value_vars=["Ledig", "Verheiratet", "Verwitwet", "Geschieden", "Lebenspartnerschaft"],
    var_name="Familienstand",
    value_name="Anzahl der Bürger"
)

print(df_melted)
```

```python
df_melted
```

```python
# Convert "Anzahl der Bürger" to numeric, replacing – with 0
df_melted["Anzahl der Bürger"] = pd.to_numeric(df_melted["Anzahl der Bürger"], errors="coerce").fillna(0)
```

```python
# Rename German columns
df_melted.rename(
    columns={
        "Altersgruppen": "Age Group",
        "Deutsche": "German",
        "Familienstand": "Marital status",
        "Anzahl der Bürger": "Number of Citizens"
    },
    inplace=True
)

print(df_melted.head())
```

```python
# Dictionary for mapping German values to English
value_mapping = {
    "Ledig": "Single",
    "Verheiratet": "Married",
    "Verwitwet": "Widowed",
    "Divorced": "Divorced",
    "Lebenspartnerschaft": "Civil Partnership",
    "65 und mehr": "65 or older"
}

# Apply
df_melted.replace(value_mapping, inplace=True)

print(df_melted.head())
```

```python
# Export to CSV
df_melted.to_csv("berlin_demographics.csv", index=False, encoding="utf-8")
```
